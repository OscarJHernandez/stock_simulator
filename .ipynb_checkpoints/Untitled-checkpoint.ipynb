```python
import numpy as np
import matplotlib.pyplot as plt
import stock_sims as st
import pandas as pd
import stochastic_diff as sd
```

```python
df = pd.read_csv("GSPC.csv")

s1 = [df["Close"][i] for i in range(len(df))]
r1 = []
log_r1 = []

for i in range(1,len(s1)):
    dr1 = (s1[i]/s1[i-1]-1.0)
    dlog_r1 = np.log(s1[i]/s1[i-1])
    r1.append(dr1)
    log_r1.append(dlog_r1)

r1 = np.asarray(r1)
print(r1.mean(),r1.std()/np.sqrt(len(r1)))
print(r1.std())

plt.plot(s1[6000:6500])
plt.show()

plt.plot(r1[6000:6500])
plt.show()

plt.hist(log_r1 ,bins=100)
plt.xlim(-0.05,0.05)
plt.show()

#plt.hist(r1[6000:6500])
plt.hist(log_r1 ,bins=100)
plt.xlim(-0.05,0.05)
plt.show()
```

```python
T =50
Tk = 1365
N_samples = 30

vN = []
VN50 = []



v1_mean = np.zeros(T+1)
v2_mean = np.zeros(T+1)
    
for i in range(0,800):
    mu_sample =  r1[Tk-N_samples:Tk].mean()
    sigma = r1[Tk-N_samples:Tk].std()

    
    dmu_sample = sigma/np.sqrt(N_samples)
    
    mu = np.random.normal(loc=mu_sample,scale=dmu_sample)
    
    S1=st.stock_simulator(mu=mu,sigma=sigma)
    S2=st.stock_simulator(mu=mu,sigma=sigma)
    
    v1=S1.simulate_stock(steps=T)
    v2=S2.simulate_GBM(steps=T)
    
    v1_mean+=v1
    v2_mean+=v2
    
    plt.plot(v2*s1[Tk],color="red",alpha=0.1)
    plt.plot(v1*s1[Tk],color="green",alpha=0.1)

v1_mean=v1_mean/800.0
v2_mean=v2_mean/800.0

plt.plot(v1_mean*s1[Tk],color="green",alpha=1.0)
plt.plot(s1[Tk:T+Tk],"-o",color="black")
plt.show()
print(mu,sigma)
```

```python
for i in range(0,500):
    mu_sample =  r1[Tk-N_samples:Tk].mean()
    sigma = r1[Tk-N_samples:Tk].std()
    dmu_sample = sigma/np.sqrt(N_samples)
    mu = np.random.normal(loc=mu_sample,scale=dmu_sample)
    
    S1=st.stock_simulator(mu=mu,sigma=sigma)
    v1=S1.simulate_stock(steps=T)
    rn = S1.stock_returns(N=1)
    plt.plot(rn,alpha=0.1,color="green")


plt.plot(r1[Tk:Tk+T],color="black")
print(r1[Tk:Tk+T].mean())
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=100)

# Fit to data using Maximum Likelihood Estimation of the parameters
x = np.asarray(range(len(r1[Tk:Tk+T]))).reshape(-1, 1)
y = np.asarray(r1[Tk:Tk+T].reshape(-1, 1))
gp.fit(x, y)
```
